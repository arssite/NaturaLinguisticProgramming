```python
doc="""JPMorgan Chase, the largest bank in the United States and the world's sixth-largest bank by total assets, is a multinational investment bank and financial services holding company headquartered in New York City. Founded in 1799 as the Bank of Manhattan Company, it merged with J.P. Morgan & Co. in 1955 and then with Chase Manhattan Corporation in 1995 to form JPMorgan Chase & Co. The company offers a wide range of financial services, including investment banking, asset management, retail banking, commercial banking, credit cards, mortgages, auto loans, and student loans."""
```

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
#nltk.word_tokenize(doc)
token=nltk.word_tokenize(doc)
```

```python
#nltk.pos_tag(token)
pos=nltk.pos_tag(token)
```
